```python
import pokemon as pk

import pandas as pd
import matplotlib.pyplot as plt 
import math
```

```python
pokemon1 = ('Pikachu',5,['Thundershock','Growl'])
pokemon2 = ('Eevee',5,['Tackle','Tail Whip'])
pokemons = (pokemon1,pokemon2)
```

```python
num_samples = 10
battles = [100,1000,5000,10000,100000]
```

```python
battle_nums = []
weedle_wins = []
eevee_wins = []
for battle_num in battles:
    for _ in range(num_samples):
        score = pk.battles(pokemons,battle_num,normalize=False)
        weedle_wins.append(score['Pikachu'])
        eevee_wins.append(score['Eevee'])
        
    battle_nums += [battle_num]*num_samples
    
```

```python
df=pd.DataFrame(data={'PikachuWins':weedle_wins,'EeveeWins':eevee_wins,"Battles":battle_nums})
df
```

```python
df['EeveeDifference'] = df['EeveeWins'] - df['PikachuWins']
```

```python
df['EeveeRatio'] = df['EeveeWins'] / df['PikachuWins']
```

```python
df['lnEeveeRatio'] = df['EeveeRatio'].map(math.log)
```

```python
plt.scatter(df.EeveeDifference,df.lnEeveeRatio)
```
